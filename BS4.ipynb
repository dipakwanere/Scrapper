```python
def scrape_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')

    # Extract data from the page
    # Modify the code below to suit your specific needs
    data = []
    for item in soup.find_all('div', class_='your-item-class'):
        # Extract relevant data from the item
        # Example:
        title = item.find('h2').text.strip()
        description = item.find('p').text.strip()
        data.append({'Title': title, 'Description': description})

    return data
```

```python
# Function to iterate through multiple pages and scrape data
def scrape_multiple_pages(base_url, num_pages):
    all_data = []
    for page_num in range(1, num_pages + 1):
        url = f"{base_url}/page/{page_num}/"
        page_data = scrape_page(url)
        all_data.extend(page_data)

    return all_data
```

```python
if __name__ == '__main__':
    base_url = 'https://localflowers.org/find-flowers'
    num_pages_to_scrape = 5  # Adjust the number of pages you want to scrape

    scraped_data = scrape_multiple_pages(base_url, num_pages_to_scrape)

    # Now, you can work with the scraped data (e.g., save to a file or process it further)
    for item in scraped_data:
        print(item)  # Print the data for demonstration
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

data = []

for i in range(1,5):    
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"
    response = requests.get(url)
    response = response.content
    soup = BeautifulSoup(response, 'html.parser')
    ol = soup.find('ol')
    articles_main = ol.find_all('article', class_ = 'product_pod') 

    for art in articles_main:
        image = art.find('img')
        title = image.attrs['alt']
        star1 = art.find('p')
        star = star1['class'][1]
        price1 = art.find('p', class_='price_color').text
        data.append([title, price1,star])
```

```python
data
```

```python
df = pd.DataFrame(data,columns=['Title','Price','Star_Rating'])
```
